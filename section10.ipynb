### Exercise 92:

###### Two files are attached to this exercise:

###### - `data_train.csv`
###### - `target_train.csv`

###### The `data_train.csv` file contains e-mails from two categories: **computer graphics** (`comp.graphics`) and **space** (`sci.space`).

###### The `target_train.csv` file contains the labels:

###### - `0` â†’ `comp.graphics`
###### - `1` â†’ `sci.space`

###### These files were loaded into the following DataFrames:

###### - `data_train`
###### - `target_train`

###### Some processing of `data_train` and `target_train` was performed.

###### The `TfidfVectorizer` class from the `scikit-learn` package was used to vectorize the text. The result was assigned to:

###### - `data_train_vectorized`

###### Using the `MultinomialNB` class, create a text document classification model. Train the model based on:

###### - `data_train_vectorized`
###### - `target_train`

###### Then classify the following sentences:

###### - `"The graphic designer requires a good processor to work."`
###### - `"Flights into space."`

###### In response, print the result to the console.

````python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

data_train = pd.read_csv('data_train.csv')
target_train = pd.read_csv('target_train.csv')

categories = ['comp.graphics', 'sci.space']

data_train = data_train['text'].to_list()
target_train = target_train.values.ravel()

vectorizer = TfidfVectorizer()
data_train_vectorized = vectorizer.fit_transform(data_train)

model = MultinomialNB()
model.fit(data_train_vectorized, target_train)

docs = ["The graphic designer requires a good processor to work", 
        "Flights into space"]

new_data = vectorizer.transform(docs)
data_pred = model.predict(new_data)

for doc, category in zip(docs, data_pred):
    print(f'\'{doc}\' => {categories[category]}') 
````

###### The `load_boston()` function from the `scikit-learn` package was used to load the data into the variable:

###### - `raw_data`

###### Then, based on the `'data'` and `'target'` keys of the `raw_data` object, the dataset was prepared.

###### The `df` object was copied into the variable:

###### - `data`

###### Then the `target` column was popped out from the `data` object and assigned to:

###### - `target`

###### Using the `train_test_split()` function, the data (`data`, `target`) was split into training and test sets and assigned to the following variables:

###### - `data_train`, `target_train`
###### - `data_test`, `target_test`

###### Using the `LinearRegression` class (with default parameters) from the `scikit-learn` package:

###### 1. Create a linear regression model.
###### 2. Fit the model on the training data.
###### 3. Evaluate the model on the test data.

###### In response, print the **RÂ˛ score** of this model to the console as shown below.

````python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression


pd.set_option('display.max_columns', 15)
pd.set_option('display.width', 150)
raw_data = load_boston()

df = pd.DataFrame(data=np.c_[raw_data.data, raw_data.target], columns=list(raw_data.feature_names) + ['target'])

data = df.copy()
target = data.pop('target')

data_train, data_test, target_train, target_test = train_test_split(data, target, random_state=42)

linear_reg = LinearRegression()
linear_reg.fit(data_train, target_train)

linear_reg.predict(data_test)

print(f'R^2 score: {linear_reg.score(data_test, target_test):.4f}')
````
