### Exercise 92:

###### Two files are attached to this exercise:

###### - `data_train.csv`
###### - `target_train.csv`

###### The `data_train.csv` file contains e-mails from two categories: **computer graphics** (`comp.graphics`) and **space** (`sci.space`).

###### The `target_train.csv` file contains the labels:

###### - `0` → `comp.graphics`
###### - `1` → `sci.space`

###### These files were loaded into the following DataFrames:

###### - `data_train`
###### - `target_train`

###### Some processing of `data_train` and `target_train` was performed.

###### The `TfidfVectorizer` class from the `scikit-learn` package was used to vectorize the text. The result was assigned to:

###### - `data_train_vectorized`

###### Using the `MultinomialNB` class, create a text document classification model. Train the model based on:

###### - `data_train_vectorized`
###### - `target_train`

###### Then classify the following sentences:

###### - `"The graphic designer requires a good processor to work."`
###### - `"Flights into space."`

###### In response, print the result to the console.

````python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

data_train = pd.read_csv('data_train.csv')
target_train = pd.read_csv('target_train.csv')

categories = ['comp.graphics', 'sci.space']

data_train = data_train['text'].to_list()
target_train = target_train.values.ravel()

vectorizer = TfidfVectorizer()
data_train_vectorized = vectorizer.fit_transform(data_train)

model = MultinomialNB()
model.fit(data_train_vectorized, target_train)

docs = ["The graphic designer requires a good processor to work", 
        "Flights into space"]

new_data = vectorizer.transform(docs)
data_pred = model.predict(new_data)

for doc, category in zip(docs, data_pred):
    print(f'\'{doc}\' => {categories[category]}') 
````
