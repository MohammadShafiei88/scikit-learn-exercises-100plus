###### Two files are attached to this exercise:

###### - `data_train.csv`
###### - `target_train.csv`

###### The `data_train.csv` file contains e-mails from two categories: **computer graphics** (`comp.graphics`) and **space** (`sci.space`).

###### The `target_train.csv` file contains the labels:

###### - `0` → `comp.graphics`
###### - `1` → `sci.space`

###### These files were loaded into the following DataFrames:

###### - `data_train`
###### - `target_train`

###### Some processing of `data_train` and `target_train` was performed.

###### Using the `TfidfVectorizer` class from the `scikit-learn` package, vectorize the text from the `data_train` list and assign it to the variable:

###### - `data_train_vectorized`

###### In response, print the shape of the resulting sparse matrix to the console.

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

data_train = pd.read_csv('data_train.csv')
target_train = pd.read_csv('target_train.csv')

categories = ['comp.graphics', 'sci.space']

data_train = data_train['text'].tolist()
target_train = target_train.values.ravel()

vectorizer = TfidfVectorizer()
data_train_vectorized = vectorizer.fit_transform(data_train)

print(data_train_vectorized.shape)

```
