###### Using the `IsolationForest` class from the `scikit-learn` package, analyze the outliers in this dataset.

###### Set the following arguments:

###### - `n_estimators=100`
###### - `contamination=0.05`
###### - `random_state=42`

###### Reminder:

###### - `1` → normal observation
###### - `-1` → outlier

###### Assign a new column to the `df` DataFrame named `'outlier_flag'` that stores information whether a given sample is normal or an outlier.

###### In response, print the first ten rows of the `df` DataFrame to the console.

````python
import pandas as pd
from sklearn.ensemble import IsolationForest

df = pd.read_csv('factory.csv')

model = IsolationForest(n_estimators=100, contamination=0.05, random_state=42)

y_pred = model.fit_predict(df)

df['outlier_flag'] = y_pred

print(df.head(10))
````
