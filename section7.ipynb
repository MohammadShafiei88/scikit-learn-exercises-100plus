### Exercise 61:

###### Load the `clusters.csv` file into a DataFrame.

###### Using the `KMeans` class from the `scikit-learn` package (`random_state=42`), create a list of WCSS (*Within-Cluster Sum-of-Squares*) values for the number of clusters from `2` to `9` inclusive.

###### Requirements:

###### - calculate WCSS for each number of clusters
###### - round WCSS values to two decimal places

###### In response, print the result to the console.

```python
import numpy as np
import pandas as pd

from sklearn.cluster import KMeans

df = pd.read_csv('clusters.csv')

wcss = []
for num_clus in range(2,10):
    model = KMeans(n_clusters=num_clus, random_state=42)
    model.fit(df)
    wcss.append(float(np.round(model.inertia_,2)))

print(wcss)
```

### Exercise 62:

###### Load the `clusters.csv` file into a DataFrame.

###### The distribution of the variables from this file is shown below.

###### Using the `KMeans` class from the `scikit-learn` package, a list of WCSS (Within-Cluster Sum-of-Squares) values was calculated.

###### Using the elbow method, select the appropriate number of clusters.

###### In response, print the selected number of clusters to the console.

```python
import numpy as np
import pandas as pd

from sklearn.cluster import KMeans
np.random.seed(42)

df = pd.read_csv('clusters.csv')

wcss = []
for num_clus in range(2,30):
    model = KMeans(n_clusters=num_clus, random_state=42)
    model.fit(df)
    wcss.append(round(float(model.inertia_),2))

print(wcss)
```

### Exercise 63:

###### Load the `clusters.csv` file into a DataFrame and assign it to the `df` variable.

###### Using the `AgglomerativeClustering` class from the `scikit-learn` package, create a model to split the given dataset into two clusters.

###### Make predictions based on this model and assign a new column `'cluster'` which stores the cluster number for each sample in the `df` DataFrame.

### Exercise 64:

###### Load the `clusters.csv` file into a DataFrame and assign it to the `df` variable.

###### Using the `AgglomerativeClustering` class from the `scikit-learn` package, create a model to split the given dataset into two clusters.

###### Use the Manhattan metric.

###### Make predictions based on this model and assign a new column `'cluster'` which stores the cluster number for each sample in the `df` DataFrame.

###### In response, print the first ten rows of the `df` DataFrame to the console.

```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering

df = pd.read_csv('clusters.csv')
 
cluster = AgglomerativeClustering(
    n_clusters=2, 
    metric='manhattan', 
    linkage='complete'
)
cluster.fit_predict(df)
 
df = pd.DataFrame(df, columns=['x1', 'x2'])
df['cluster'] = cluster.labels_
 
print(df.head(10))
```

### Exercise 65:

###### Load the `clusters.csv` file into a DataFrame and assign it to the `df` variable.

###### Using the `DBSCAN` class from the `scikit-learn` package, create a model to split the given dataset into clusters.

###### Set the following arguments:

###### - `eps=0.6`
###### - `min_samples=7`

###### Make predictions based on this model and assign a new column `'cluster'` which stores the cluster number for each sample in the `df` DataFrame.

###### In response, print the first ten rows of the `df` DataFrame.

```python
import pandas as pd
from sklearn.cluster import DBSCAN

df = pd.read_csv('clusters.csv')

model = DBSCAN(eps=0.6, min_samples=7)

df['cluster'] = model.fit_predict(df)
print(df.head(10))
```
