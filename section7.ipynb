### Exercise 61:

###### Load the `clusters.csv` file into a DataFrame.

###### Using the `KMeans` class from the `scikit-learn` package (`random_state=42`), create a list of WCSS (*Within-Cluster Sum-of-Squares*) values for the number of clusters from `2` to `9` inclusive.

###### Requirements:

###### - calculate WCSS for each number of clusters
###### - round WCSS values to two decimal places

###### In response, print the result to the console.

```python
import numpy as np
import pandas as pd

from sklearn.cluster import KMeans

df = pd.read_csv('clusters.csv')

wcss = []
for num_clus in range(2,10):
    model = KMeans(n_clusters=num_clus, random_state=42)
    model.fit(df)
    wcss.append(float(np.round(model.inertia_,2)))

print(wcss)
```
