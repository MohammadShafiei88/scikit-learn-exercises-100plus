### Exercise 61:

###### Load the `clusters.csv` file into a DataFrame.

###### Using the `KMeans` class from the `scikit-learn` package (`random_state=42`), create a list of WCSS (*Within-Cluster Sum-of-Squares*) values for the number of clusters from `2` to `9` inclusive.

###### Requirements:

###### - calculate WCSS for each number of clusters
###### - round WCSS values to two decimal places

###### In response, print the result to the console.

```python
import numpy as np
import pandas as pd

from sklearn.cluster import KMeans

df = pd.read_csv('clusters.csv')

wcss = []
for num_clus in range(2,10):
    model = KMeans(n_clusters=num_clus, random_state=42)
    model.fit(df)
    wcss.append(float(np.round(model.inertia_,2)))

print(wcss)
```

### Exercise 68:

###### The `pca.csv` file was loaded into the `df` DataFrame. Using the `StandardScaler` class, the variables in `X` were standardized and assigned to `X_std`.

###### Implement the PCA algorithm using the `X_std` array. Reduce the result to two principal components and assign it to `X_pca`.

###### In response, print the first ten rows of the `X_pca` array.

###### Steps:

###### 1. Compute the covariance matrix of the `X_std` array  
###### 2. Find the eigenvectors and their corresponding eigenvalues for this covariance matrix  
###### 3. Sort eigenvectors by decreasing eigenvalues  
###### 4. Determine the number of components (in this case 2)  
###### 5. Create matrix `W` from selected vectors (columns as eigenvectors)  
###### 6. Multiply `X_std` by `W` and assign to the `X_pca` variable

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler


np.random.seed(42)
df = pd.read_csv('pca.csv')

X = df.copy()
y = X.pop('class')

scaler = StandardScaler()
X_std = scaler.fit_transform(X)

eig_vals, eig_vecs = np.linalg.eig(np.cov(X_std, rowvar=False))
eig_pairs = [
    (np.abs(eig_vals[i]), eig_vecs[:, i]) 
    for i in range(len(eig_vals))
]
eig_pairs.sort(reverse=True)
 
W = np.hstack((
    eig_pairs[0][1].reshape(3, 1), 
    eig_pairs[1][1].reshape(3, 1)
))
X_pca = X_std.dot(W)
print(X_pca[:10])
```

### Exercise 69:

###### The PCA algorithm was implemented using the `X_std` array and the result was assigned to the `X_pca` variable.

###### Create a DataFrame called `df_pca` using the `X_pca` array and the `y` variable.

###### In response, print the first ten rows of the `df_pca` object to the console.

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler

np.random.seed(42)

df = pd.read_csv('pca.csv')
X = df.drop('class', axis=1)
y = df['class']

scaler = StandardScaler()
X_std = scaler.fit_transform(X)

eig_vals, eig_vecs = np.linalg.eig(np.cov(X_std,rowvar=False))
eig_pairs = [ (np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range (len(eig_vals))]
eig_pairs.sort(reverse=True)

W = np.hstack((eig_pairs[0][1].reshape(3,1),eig_pairs[1][1].reshape(3,1)))
X_pca = X_std.dot(W)

df_pca = pd.DataFrame(X_pca, columns=['pca_1', 'pca_2'])
df_pca['pca_2'] = - df_pca['pca_2']
df_pca ['class'] = y
print(df_pca.head(10))
```
