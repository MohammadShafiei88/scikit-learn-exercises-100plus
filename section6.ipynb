### Exercise 51:

###### The following arrays are given:

###### - `X_train`, `y_train`  
###### - `X_test`, `y_test`  

###### Using the `RandomForestClassifier` class from the `scikit-learn` package, create a classification model (`random_state=42`).

###### Train the model on the training set and evaluate it on the test set.

###### In response, print the model accuracy (up to four decimal places) to the console as shown below.

````python
import numpy as np
import pandas as pd

from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier


np.random.seed(42)
raw_data = make_moons(n_samples=2000, noise=0.25, random_state=42)
data = raw_data[0]
target = raw_data[1]

X_train, X_test, y_train, y_test = train_test_split(data, target)

model = RandomForestClassifier(random_state=42)

model.fit(X_train, y_train)
print(f'Accuracy: {model.score(X_test,y_test):.4f}')
````

### Exercise 52:

###### The following arrays are given:

###### - `X_train`, `y_train`  
###### - `X_test`, `y_test`  

###### Using the `RandomForestClassifier` class and grid search method (`GridSearchCV` with `scoring='accuracy'`, `cv=2`), find the optimal values of the following parameters:

###### - `criterion` → `['gini', 'entropy']`
###### - `max_depth` → `[6, 7, 8]`
###### - `min_samples_leaf` → `[4, 5]`

###### Train the model on the training set and evaluate it on the test set.

###### In response, print the optimal values of `criterion`, `max_depth`, and `min_samples_leaf` parameters to the console.

````python
import numpy as np
import pandas as pd

from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

np.random.seed(42)
raw_data = make_moons(n_samples=2000, noise=0.25, random_state=42)
data = raw_data[0]
target = raw_data[1]

X_train, X_test, y_train, y_test = train_test_split(data, target)

params = {'criterion': ['gini', 'entropy'],
          'max_depth':[6, 7, 8],
          'min_samples_leaf': [4,5]}

model = RandomForestClassifier(random_state=42)
grid_cv = GridSearchCV(
    model,
    param_grid=params,
    scoring='accuracy',
    cv=2
)

grid_cv.fit(X_train,y_train)

grid_cv.score(X_test, y_test)

print(grid_cv.best_params_)
````

### Exercise 53:

###### The following list with text documents is given:

```python
documents = [
    'python is a programming language',
    'python is popular',
    'programming in python',
    'object-oriented programming in python'
]
```

###### Vectorize the documents using the `CountVectorizer` class from the `scikit-learn` package.

###### In response, print the result as a DataFrame.

````python
import numpy as np
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer


documents = [
    'python is a programming language',
    'python is popular',
    'programming in python',
    'object-oriented programming in python'
]

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

col_name = vectorizer.get_feature_names_out()
data = X.toarray()

df = pd.DataFrame(data=data, columns=col_name)
print(df)
````
