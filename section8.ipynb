###### Load the `pca.csv` file into the `df` DataFrame.

###### Perform dimensionality reduction using PCA with 3 principal components using the `PCA` class from the `scikit-learn` package.

###### In response, print the percentage of variance explained by these components as a DataFrame object.

````python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = pd.read_csv('pca_1.csv')

scaler = StandardScaler()

df_std = scaler.fit_transform(df)

pca = PCA(n_components=3)
df_pca = pca.fit_transform(df_std)

results = pd.DataFrame(
    data={'explained_variance_ratio': pca.explained_variance_ratio_}
)
results['cumulative'] = results['explained_variance_ratio'].cumsum()
results['component'] = results.index + 1
print(results)
````

###### One-hot encoding was used to encode the transaction database (`trans_encoded_df` variable):

   bread  butter  cheese  eggs  milk

###### Calculate the support for individual products and present the result as a DataFrame object.

###### In response, print this DataFrame to the console.

###### > Tip:  
###### > support(product A) = number of transactions containing product A / total number of transactions in the database

````python
import pandas as pd
import numpy as np

data = {
    "products": [
        "bread eggs",
        "bread eggs milk",
        "milk cheese",
        "bread butter cheese",
        "eggs milk",
        "bread milk butter cheese",
    ]
}

transactions = pd.DataFrame(data=data, index = range(1,7))
expanded = transactions['products'].str.split(' ', expand=True)

products = []

for idx, row in expanded.iterrows():
    for item in row.dropna():
        if item not in products:
            products.append(item)

products.sort()

trans_encoded_df = pd.DataFrame(0, index=expanded.index, columns=products)

for idx,row in expanded.iterrows():
    for item in row.dropna():
        trans_encoded_df.loc[idx,item]=1

support = trans_encoded_df.sum()/len(trans_encoded_df)
print(support)

````

###### Calculate the support for the following pairs:

###### - `(butter, bread)`
###### - `(butter, milk)`

###### In response, print the result to the console.

###### Round the support values to four decimal places.

###### > Tip:  
###### > support(product A, product B) = number of transactions containing product A and product B / total number of transactions in the database

````python
import pandas as pd
import numpy as np

data = {
    "products": [
        "bread eggs",
        "bread eggs milk",
        "milk cheese",
        "bread butter cheese",
        "eggs milk",
        "bread milk butter cheese",
    ]
}

transactions = pd.DataFrame(data=data, index = range(1,7))
expanded = transactions['products'].str.split(' ', expand=True)

products = []

for idx, row in expanded.iterrows():
    for item in row.dropna():
        if item not in products:
            products.append(item)

products.sort()

trans_encoded_df = pd.DataFrame(0, index=expanded.index, columns=products)

for idx, row in expanded.iterrows():
    for item in row.dropna():
        trans_encoded_df.loc[idx, item]=1

sup_butter_bread = len(trans_encoded_df.query("butter == 1 and bread == 1")) / len(trans_encoded_df)
sup_butter_milk = len(trans_encoded_df.query("butter == 1 and milk == 1")) / len(trans_encoded_df)
 
print(f'support(butter, bread) = {sup_butter_bread:.4f}')
print(f'support(butter, milk) = {sup_butter_milk:.4f}')
````
