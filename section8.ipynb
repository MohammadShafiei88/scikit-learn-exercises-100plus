### Exercise 71:

###### Load the `pca.csv` file into the `df` DataFrame.

###### Perform dimensionality reduction using PCA with 3 principal components using the `PCA` class from the `scikit-learn` package.

###### In response, print the percentage of variance explained by these components as a DataFrame object.

````python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

df = pd.read_csv('pca_1.csv')

scaler = StandardScaler()

df_std = scaler.fit_transform(df)

pca = PCA(n_components=3)
df_pca = pca.fit_transform(df_std)

results = pd.DataFrame(
    data={'explained_variance_ratio': pca.explained_variance_ratio_}
)
results['cumulative'] = results['explained_variance_ratio'].cumsum()
results['component'] = results.index + 1
print(results)
````

### Exercise 74:

###### A `transactions` DataFrame is given:

```python
products
1                bread eggs
2           bread eggs milk
3               milk cheese
4       bread butter cheese
5                 eggs milk
6  bread milk butter cheese
```

###### Each row contains products purchased by one customer.

###### Each row of the `products` column was split by a space character and expanded into a DataFrame (stored in the variable `expanded`).

###### This step has already been completed.

###### In response, use the `expanded` variable to continue the task as shown below.

###### Extract all unique product names from the `expanded` DataFrame and assign them to the `products` variable (sorted alphabetically).

###### In response, print the `products` list to the console.

````python
import numpy as np
import pandas as pd


data = {
    "products": [
        "bread eggs",
        "bread eggs milk",
        "milk cheese",
        "bread butter cheese",
        "eggs milk",
        "bread milk butter cheese",
    ]
}

df = pd.DataFrame(data=data, index=range(1,7))
expanded = df['products'].str.split(expand=True)

products = []

for col in expanded.columns:
    for item in expanded[col].unique():
        if type(item) is not float and item not in products:
            products.append(item)

products.sort()

print(products)

````
