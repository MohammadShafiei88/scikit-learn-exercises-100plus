### Exercise 11:

###### The `df` DataFrame is given below:

|    | weight | weight_cut |
|---:|-------:|:----------:|
| 0 | 75.0 | light |
| 1 | 78.5 | normal |
| 2 | 85.0 | heavy |
| 3 | 91.0 | heavy |
| 4 | 84.5 | heavy |
| 5 | 83.0 | heavy |
| 6 | 68.0 | light |

###### Using the `pd.get_dummies()` function convert the categorical column `weight_cut` into dummy variables.

###### In response, print the result to the console.

```python
import pandas as pd

df = pd.DataFrame(data={'weight': [75., 78.5, 85., 91., 84.5, 83., 68.]})
df['weight_cut'] = pd.cut(df['weight'], bins=(60, 75, 80, 95), labels=['light', 'normal', 'heavy'])

df = pd.get_dummies(df, dtype=int)
print(df)
```

### Exercise 12

###### The `df` DataFrame is given below:

|    | currency |
|---:|:---------------------------|
| 0 | [PLN, USD] |
| 1 | [EUR, USD, PLN, CAD] |
| 2 | [GBP] |
| 3 | [JPY, CZK, HUF] |
| 4 | [] |

###### Assign a new column `'number'` that takes the number of items in the `currency` column.

###### In response, print the `df` DataFrame to the console.

```python
import pandas as pd

data_dict = {
    'currency': [['PLN', 'USD'], ['EUR', 'USD', 'PLN', 'CAD'], ['GBP'], ['JPY', 'CZK', 'HUF'], []]
}
df = pd.DataFrame(data=data_dict)
df['number'] = df['currency'].apply(len)
print(df)
```

### Exercise 13

###### The `df` DataFrame is given below:

|    | currency |
|---:|:---------------------------|
| 0 | [PLN, USD] |
| 1 | [EUR, USD, PLN, CAD] |
| 2 | [GBP] |
| 3 | [JPY, CZK, HUF] |
| 4 | [] |

###### Assign a new column `'PLN_flag'` which stores `1` when the currency `'PLN'` is in the `currency` column, otherwise `0` as shown below.

###### In response, print the `df` DataFrame to the console.

```python
import pandas as pd

data_dict = {
    'currency': [['PLN', 'USD'], ['EUR', 'USD', 'PLN', 'CAD'], ['GBP'], ['JPY', 'CZK', 'HUF'], []]
}

df = pd.DataFrame(data = data_dict)

df['PLN_flag'] = df['currency'].apply(lambda i:1 if 'PLN' in i else 0)
print(df)
```
