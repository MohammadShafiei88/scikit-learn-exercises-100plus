### Exercise 44:

###### Implement the `entropy()` function using the `numpy` package.

###### Round the result to four decimal places.

###### Then the following `df` DataFrame is given:

|    | val_1 | val_2 |
|---:|------:|------:|
| 0 | 0.01 | 0.99 |
| 1 | 0.11 | 0.89 |
| 2 | 0.21 | 0.79 |
| 3 | 0.31 | 0.69 |
| 4 | 0.41 | 0.59 |
| 5 | 0.51 | 0.49 |
| 6 | 0.61 | 0.39 |
| 7 | 0.71 | 0.29 |
| 8 | 0.81 | 0.19 |
| 9 | 0.91 | 0.09 |

###### Using the implemented function, calculate the entropy for each pair in this DataFrame and assign it to the `'entropy'` column.

###### In response, print the `df` DataFrame to the console.

```python
import numpy as np
import pandas as pd

data_dict = {
    'val_1': np.arange(0.01, 1.0, 0.1), 
    'val_2': 1 - np.arange(0.01, 1.0, 0.1)
}
df = pd.DataFrame(data_dict)

def entropy(x):
    return np.round(-np.sum(x*np.log2(x)),4)

df['entropy'] = df.apply(entropy, axis=1)
print(df)
```

### Exercise 47:

###### The following arrays are given:

###### - `X_train`, `y_train`
###### - `X_test`, `y_test`

###### Using the `DecisionTreeClassifier` class from the `scikit-learn` package, create a classification model (leave the default arguments).

###### Train the model on the training set and evaluate it on the test set.

###### In response, print the model accuracy (up to four decimal places) to the console as shown below.

```python
import numpy as np
import pandas as pd

from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier


np.random.seed(42)
raw_data = make_moons(n_samples=2000, noise=0.25, random_state=42)
data = raw_data[0]
target = raw_data[1]

X_train, X_test, y_train, y_test = train_test_split(data, target)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

print(f'Accuracy: {model.score(X_test, y_test):.4f}')
```
