###### Using the implemented function, calculate sigmoid values for the variable `var1` and assign them to the new column `'var1_sigmoid'`.

###### In response, print the `df` DataFrame to the console.

```python
import numpy as np
import pandas as pd

np.random.seed(42)
df = pd.DataFrame(data=np.random.randn(10), columns=['var1'])

df['var1_sigmoid'] = 1/(1+np.exp(-df['var1']))
print(df)
```
