### Exercise 31:

###### The Breast Cancer dataset was loaded into the `data` and `target` variables.

###### Using the `train_test_split()` function from the `scikit-learn` package, split the data (`data` and `target`) into train and test sets:

###### - train set: `X_train`, `y_train`
###### - test set: `X_test`, `y_test`

###### Set the following parameters:

###### - `random_state=40`
###### - test set size = `25%`

###### In response, print the shapes of these arrays to the console as shown below.

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

raw_data = load_breast_cancer()
data = raw_data['data']
target = raw_data['target']

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=40, test_size=0.25)

print(f'X_train shape {X_train.shape}')
print(f'X_test shape {X_test.shape}')
print(f'y_train shape {y_train.shape}')
print(f'y_test shape {y_test.shape}')

```

### Exercise 32:

###### The Breast Cancer dataset was loaded into the `data` and `target` variables.

###### All data was split into train and test sets:

###### - train set: `X_train`, `y_train`
###### - test set: `X_test`, `y_test`

###### Find the distribution of the `target`, `y_train`, and `y_test` arrays (in percent).

###### In response, print the result to the console as shown below.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

raw_data = load_breast_cancer()

data = raw_data['data']
target = raw_data['target']

X_train, X_test, y_train, y_test = train_test_split(
    data, target, random_state=40, test_size=0.25)

for name, array in zip(['target','y_train', 'y_test'],[target, y_train, y_test]):
    print(f'{name.ljust(7)}:{np.unique(array, return_counts=True)[1]/len(array)}')




```
