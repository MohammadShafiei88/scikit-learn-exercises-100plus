### Exercise 34:

###### The following `df` DataFrame is given:

|    | years | salary |
|---:|------:|-------:|
| 0 | 1 | 4000 |
| 1 | 2 | 4250 |
| 2 | 3 | 4500 |
| 3 | 4 | 4750 |
| 4 | 5 | 5000 |
| 5 | 6 | 5250 |

###### The first column describes the years of experience (feature variable), the second column describes the salary (target variable).

###### Using the normal equation and the `numpy` package, find the linear regression equation.

###### In response, print the result to the console as shown below.

```python
import numpy as np
import pandas as pd


df = pd.DataFrame({'years': [1, 2, 3, 4, 5, 6],
                   'salary': [4000, 4250, 4500, 4750, 5000, 5250]})


m = len(df)
 
X1 = df['years'].values
Y = df['salary'].values
 
X1 = X1.reshape(m, 1)
bias = np.ones((m, 1))
X = np.append(bias, X1, axis=1)
 
coefs = np.dot(np.linalg.inv(np.dot(X.T, X)), np.dot(X.T, Y))
print(f'Linear regression: {coefs[0]:.2f} + {coefs[1]:.2f}x')

```

| 4 | 5 | 5000 |
| 5 | 6 | 5250 |

###### The first column describes the years of experience (feature variable), the second column describes the salary (target variable).

###### Using the `scikit-learn` package and the `LinearRegression` class, find the linear regression equation for this problem.

###### In response, print the result to the console as shown below.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression


df = pd.DataFrame({'years': [1, 2, 3, 4, 5, 6],
                   'salary': [4000, 4250, 4500, 4750, 5000, 5250]})

X = df[['years']]
y=df[['salary']]

model = LinearRegression()
model.fit(X,y)

print(f'Linear regression: {model.intercept_[0]:.2f} '
      f'+ {model.coef_[0][0]:.2f}x')

```

### Exercise 40:

###### The `predictions.csv` file contains predictions for a regression model:

###### - `y_true` variable describes the actual, observed values  
###### - `y_pred` variable describes the values provided by the model  

###### Load this file into a DataFrame.

###### Then implement a function called `mean_squared_error()` that computes the Mean Squared Error (MSE).

###### Using this function, calculate the MSE value for this model and print the result to the console.

```python
import numpy as np
import pandas as pd

df = pd.read_csv('prediction.csv')

def mse(y_true,y_pred):
    return ((y_true - y_pred)**2).sum()/ len(y_true)
    

MSE = mse(df['y_true'],df['y_pred'])
print(f'MSE = {MSE:.4f}')

```
